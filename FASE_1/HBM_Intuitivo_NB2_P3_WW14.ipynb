## 2.3 Teorema de Bayes
Una de las primeras tareas para acostumbrarse a un nuevo framework viniendo de uno tradicional es todos los modelos se centra en la distribución de parámetros en lugar de las estimaciones individuales de ellos. En el contexto bayesiano, los parámetros son aleatorios, no fijos. Su análisis comenzará con una distribución, la anterior, y finalizará con otra, la posterior. Los resúmenes de esa distribución, p. la media, la desviación estándar, etc. estarán disponibles y, por lo tanto, se utilizarán para comprender los parámetros de manera similar al enfoque tradicional. Lo que diferencia un modelo jerárquico bayesiano es que sus parámetros de la distrivución previa dependen directamente de otros parámetros con distribuciones similares.
Los modelos Bayesianos responden específicamente al **Teorema de Bayes** el cual es el siguiente:

    
<img src="img/bayesiantheorem2.png" alt="Teorema de Bayes" class="center" width="500" height="600" >


### 2.3.1 P(θ) Distribución Previa --> P(H)
Las prioridades a priori son las que se consideran independientes de algún evento, es decir que se conoce el resultado de la afirmación antes de que el evento sea investigado o indagado. Estas indican que tan probable es que suceda un evento en la naturaleza.

Para el enfoque bayesiano debemos elegir una **distribución previa** que represente lo que creemos sobre la estimación. Esto es tradicionalmente donde algunos específicamente tienen dificultades con la estimación bayesiana, y generalmente cuesta más entender el modelo por su naturaleza. Es aquí donde se deben tomar decisiones de que modelo utilizar, pero son muy similares a las ya antes tomados en modelos estadísticos generales. No se debe dejar que un programa realice este aproximación por uno. Los ejemplos incluyen qué variables entran en el modelo, la naturaleza de la probabilidad (por ejemplo, distribucion noraml vs Poisson), si incluir interacciones o no, etc. Siempre se está haciendo este tipo de elecciones en el modelado estadístico.

La configuración anterior se basa generalmente en preocupaciones de modelado. Como ejemplo, el coeficiente de regresión anterior podría establecerse de manera que limite nuestro modelado y le de un sentido práctico al modelo. Esto iría en contra del propósito del enfoque bayesiano, pero representa una situación que ignoramos por completo. De hecho, hacerlo produciría los resultados de los enfoques de regresión de probabilidad estándar y, por lo tanto, puede pensar en su enfoque familiar como uno bayesiano con antecedentes no informados. Sin embargo, la práctica más común generalmente establece lo anterior para que un coeficiente de regresión sea normal, con media en cero y con una varianza relativamente grande.

### 2.3.2 P(θ|x) Distribución Posterior --> P(H|E)

Son las **probabilidades a posteriori**, las cuales son aquellas que dependen totalmente de que un evento suceda, es decir no se sabe su resultado antes del evento. θ Theta representa los parámetros del modelo que pueden producir datos y el símbolo | --> dado. Esta distribución es una combinación ponderada de la priori y la muestral, y es proporcional al producto de ambas.

### 2.3.3 P(x|θ) Distribución Muestral --> P(E|H)
La **distribución muestral** liga el modelo con los datos probabilisticamente, que tan probables son los datos dados específicos Thetas. Para entender un poco mejor el concepto de distribución muestral, y probabilidad se utilizan las siguientes imágenes.

La primera gráfica representa la probabilidad de que un evento suceda dado ciertas condiciones, se calcula encontrando el área bajo la curva dado una distribución continua. Por otro lado la distribución muestral se observa en la imagen de abajo, la cual dado un valor, o valores únicos θ, se sabe su valor en el eje y. La diferencia principal es que una distribución muestral esta ligada a la forma de la distribución, específicamente a la desviación estandar, si esta varía el resultado muestral va a cambiar igualmente. La distribución muestral nos indica además, dados los parámetros de estimación, cuantas veces los datos pueden ocurrir. Un factor clave que diferencia los métodos de Machine learning tradicionales con el Bayesiano, es que los primeros asumen parámetros fijos, mientras que los Bayesianos son aleatorios.

# 4. Entendiendo mejor cada Distribución
## 4.2 Ejemplo descriptivo de como se comporta cada distribución

La idea del siguiente ejemplo es entender como se genera la distribución **posterior** utilizando como base la **previa** y **likelihood**. Es importante entender que esta distribución posterior se considera en cierta manera más precisa ya que considera y toma en cuenta en sus predicciones la distribución previa.\
Para el ejemplo nos vamos a imaginar que contamos con una cuenta de youtube en donde diariamente se publican videos, los cuales son vistos por 2000 personas diariamente. Algunos de estas personas le dan like al video después de haberlo visto y algunos no. La idea es predecir el porcentaje de persoas que le dan like despues de haber visto el video.

```python
import numpy as np

# Se limita el número de elemntos en el arreglo que serán mostrados
np.set_printoptions(threshold=100)

# A continuación se genera la simulación de la respuesta de 2000 personas  
# En este ejemplo 1 = like, 0 = no like

# Se supone una tasa de exito de un 30%.
like_prob = 0.3

# Distribución binomial para simular las personas
like_data = np.random.binomial(n=1, p=like_prob, size=2000)
```

Como se observa a continuación el arreglo para los likes es uno aleatorio de unos y ceros y de largo 2000.

```python
like_data
```

```python
len (like_data)
```

### 4.2.1 Cálculo de la distribución previa P(θ) -> P(H)
Se debe escoger el modelo que vamos a utilizar para determinar el parámetro θ, que para nuestro caso representa la probabilidad de que una persona le de like al video. La pregunta que tenemos que hacernos es ¿Que tipo de distribución probabilística debe seguir el modelo? para esto debemos representar la probabilidad tomando en cuenta que el rango de la distribución va a ser de 0 a 1, y debe ser continua.\
Se va a escoger una distribución **Beta** ya que esta es de tipo univariante, es decir para una única variable como es nuestro caso (1 o 0).\
Para utilizar una distribución Beta se deben tomar en cuenta dos parámetros, α & β. α va a considerarse como la cantidad de personas que le dan like a los videos, y β cuantas personas no lo hacen. Vamos a suponer que el día de hoy se registraron 400 likes. Se tendría

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

#alpha
a = 400
b = 2000 - a

# se define el dominio de θ
rango_theta = np.linspace(0, 1, 1000)

# definición de la distribución previa
previa = stats.beta.pdf(x = rango_theta, a=a, b=b)
```

```python
# Plot de la distribucion previa
plt.rcParams['figure.figsize'] = [20, 7]
fig, ax = plt.subplots()
plt.plot(rango_theta, previa, linewidth=3, color='green')
# Titulo
plt.title('Probabilidad de Likes (Previa)"', fontsize=20)
plt.xlabel('θ', fontsize=16)
plt.ylabel('Densidad', fontsize=16)

plt.grid(alpha=.4, linestyle='--')

plt.show()
```

### 4.2.2 Cálculo del likelihood P(X|θ) -> P(E|H)

El cálculo de la distribución muestral esta ligado a la probabilidad de ver los datos (la evidencia) dado un dato θ en particular. Para saber que tipo de distribución utilizar se de be saber como son nuestros datos, recordemos que es un arreglo de 1s y 0s. Además sabemos cual es el número total de personas (n) y queremos la probabilidad de que estos le dan like (p), esto se considera una distribución binomial.

```python
#Distribución P(X|θ) con un único θ

likelihood = stats.binom.pmf(k = np.sum(like_data), n = len(like_data), p = a/(a+b))
```

```python
# Que ta probable es que alguien aplauda
likelihood
```

Como se puede nuestra suposición no es probable que suceda. 

```python
# Plot de la distribución muestral (likelihood)

# Likelihood P(X|θ) para todos los θ's
likelihood = stats.binom.pmf(k = np.sum(like_data), n = len(like_data), p = rango_theta)
fig, ax = plt.subplots()
plt.plot(rango_theta, likelihood, linewidth=3, color='yellow')
# Título
plt.title('Likelihood -> Probabilidad de likes' , fontsize=20)
plt.xlabel('θ', fontsize=16)
plt.ylabel('Probabilidad', fontsize=16)

plt.grid(alpha=.4, linestyle='--')

plt.show()
```

### 4.2.3 Distribución posterior P(θ|X) -> P(H|E)

La forma mas sencilla de realizar este cálculo es escogiendo el modelo correctamente, como ya vimos binario y binomial. Se calcula la distribución previa restando dos distribuciones, nótese que se utiliza **.cdf** en vez de **.pdf** esto debido a la distribución muestral **.pmf** es una probabilidad mientras que **.pdf** regresa una densidad de probabilidad. 



```python
rango_theta_e = rango_theta + 0.001 
previa = stats.beta.cdf(x = rango_theta_e, a=a, b=b) - stats.beta.cdf(x = rango_theta, a=a, b=b) 

likelihood = stats.binom.pmf(k = np.sum(like_data), n = len(like_data), p = rango_theta) 
posterior = likelihood * previa # multiplicación de las distribuciones
posterior_normalizada = posterior / np.sum(posterior)
```

```python
# Plot de las 3 distribuciones
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(20,7))
plt.xlabel('θ', fontsize=24)
axes[0].plot(rango_theta, previa, label="Previa", linewidth=3, color='green')
axes[0].set_title("Previa", fontsize=16)
axes[1].plot(rango_theta, likelihood, label="Likelihood", linewidth=3, color='yellow')
axes[1].set_title("Distribución Muestral (Likelihood)", fontsize=16)
axes[2].plot(rango_theta, posterior, label='Posterior', linewidth=3, color='gray')
axes[2].set_title("Posterior", fontsize=16)
plt.show()
```

Como se puede observar en los gráficos anteriores, la distribución muestral tiene un 30% de probabilidad de que hayan likes, utilizando la previa un 20%, mientras que si nos fijamos en el tercer gráfico este tiene un 25% de probabilidad de que haya un like y es el que el modelo Bayesiano estima. Es importante esta visualización gráfica ya que siempre esto deberia suceder, la distribución posterior es un valor entre la previa y la muestral.
